```python
import numpy as np
import pandas as pd
from sklearn.metrics import jaccard_score
from scipy.spatial.distance import euclidean,cityblock,chebyshev,hamming
from sklearn.metrics.pairwise import cosine_similarity
```

```python
customer_A = [4, 5, 2, 3, 4] 
customer_B = [5, 3, 2, 4, 5] 
```

```python
customer_A_binary = [1, 0, 1, 1, 0, 1] 
customer_B_binary = [1, 1, 1, 0, 0, 1] 
```

```python
def dist(A,B):
    A = np.array(A)
    B = np.array(B)
    print(f"1. {euclidean(A,B)=}")
    print(f"2. {cityblock(A,B)=}")
    print(f"3. {hamming(A,B)=}")
    print(f"4. {cosine_similarity(A.reshape(1,-1),B.reshape(1,-1))=}")
    print(f"5. {chebyshev(A,B)=}")
```

```python
dist(customer_A_binary,customer_B_binary)
```

```python
from scipy.spatial.distance import chebyshev, minkowski

user1 = [5, 3, 4, 4, 2]
user2 = [4, 2, 5, 4, 3]

chebyshev_dist = chebyshev(user1, user2)
minkowski_dist = minkowski(user1, user2, p=3)   # you can change p to 1, 2, etc.

print(f"Chebyshev Distance: {chebyshev_dist}")
print(f"Minkowski Distance: {minkowski_dist}")
```
