```python
# Assignment 5.2: Add all numbers divisible by 7 and 9 from 1 to n

n = int(input("Enter the value of n: "))

total = 0
for i in range(1, n + 1):
    if i % 7 == 0 and i % 9 == 0:   # divisible by both 7 and 9
        total += i

print(f"\nThe sum of all numbers divisible by both 7 and 9 from 1 to {n} is: {total}")
```

```python
# Assignment 5.3: Add all prime numbers from 1 to n

n = int(input("Enter the value of n: "))

def is_prime(num):
    if num < 2:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

total = 0
for i in range(1, n + 1):
    if is_prime(i):
        total += i

print(f"\nThe sum of all prime numbers from 1 to {n} is: {total}")
```

```python
# Assignment 6.1: Add all odd numbers from 1 to n (using function)

def add_odd_numbers(n):
    total = 0
    for i in range(1, n + 1):
        if i % 2 != 0:   # odd number check
            total += i
    return total

# --- main program ---
n = int(input("Enter the value of n: "))
result = add_odd_numbers(n)
print(f"\nThe sum of all odd numbers from 1 to {n} is: {result}")
```

```python
# Assignment 6.2: Add all prime numbers from 1 to n (using function)

def is_prime(num):
    if num < 2:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

def add_prime_numbers(n):
    total = 0
    for i in range(1, n + 1):
        if is_prime(i):
            total += i
    return total

# --- main program ---
n = int(input("Enter the value of n: "))
result = add_prime_numbers(n)
print(f"\nThe sum of all prime numbers from 1 to {n} is: {result}")
```

```python
import string as s
import random as r
print ("String        --> ",s.ascii_letters)

passwd=r.sample(s.ascii_letters, 6)
print ("Selected Char --> ",passwd)

passwd1="".join(passwd)
print ("passwd1       --> ",passwd1)

passwd2="+".join(passwd)
print ("passwd2       --> ",passwd2)

passwd3="*".join(passwd)
print ("passwd3       --> ",passwd3)

```

```python
import string as s
import random as r
print ("String + Digits --> ",s.ascii_letters + s.digits)

mixPasswd=r.sample(s.ascii_letters + s.digits, 5)
print ("\nSelected Str1 --> ",mixPasswd)
mixPasswd="".join(mixPasswd)
print ("mixPasswd1    --> ",mixPasswd)

mixPasswd=r.sample(s.ascii_letters + s.digits, 6)
print ("\nSelected Str2 --> ",mixPasswd)
mixPasswd="".join(mixPasswd)
print ("mixPasswd2    --> ",mixPasswd)

splChar="#@!~%^&*()_+=-[]{}|"
mixPasswd=r.sample(splChar + s.ascii_letters + s.digits, 8)
print ("\nSelected Str3 --> ",mixPasswd)
mixPasswd="".join(mixPasswd)
print ("mixPasswd3    --> ",mixPasswd)
     
```
